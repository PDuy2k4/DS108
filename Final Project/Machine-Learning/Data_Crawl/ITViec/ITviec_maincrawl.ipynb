```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time 
import csv
```

```python
def get_job_title(driver) -> str: 
    try: 
        title = driver.find_element(By.XPATH, "//div[@class='job-header-info']//h1")
        return title.text
    except:
        return None
```

```python
def get_company_name(driver) -> str: 
    try: 
        company_name = driver.find_element(By.XPATH, "//div[@class='job-header-info']//div")
        return company_name.text
    except:
        return None
```

```python
def get_requirement(driver) -> str: 
    reqs = []
    try: 
        requirements = driver.find_elements(By.XPATH, "/html/body/main/div[1]/div[1]/div[2]/div[1]/div[1]/section/div[5]/ul/li")
        for red in requirements:
            reqs.append(red.text)
        return reqs 
    except:    
        return None     
```

```python
def get_salary(driver) ->str: 
    return "Negotiable"
    
```

```python
def get_technical_skill(driver) -> str: 
    try:
        skill_list = []
        tech_skill = driver.find_elements(By.XPATH, "/html/body/main/div[1]/div[1]/div[2]/div[1]/div[1]/div[2]/div/div/div[4]/a")
        for tech in tech_skill:
            skill_list.append(tech.text)
        return skill_list
    except:
        return None
```

```python
def get_location(driver) -> str: 
    try:
        loc = driver.find_element(By.XPATH, "/html/body/main/div[1]/div[1]/div[2]/div[1]/div[1]/div[2]/div/div/div[1]/span")
        location = loc.text
        return location
    except:
        return None 
```

```python
def get_time() -> str: 
    return "Full time"
```

```python
def get_image_link(driver) -> str: 
    try: 
        img_link = driver.find_element(By.XPATH, "/html/body/main/div[1]/div[1]/div[2]/div[1]/div[2]/section/div[1]/a/picture/img")
        img = img_link.get_attribute("data-src")
        return img 
    except: 
        return None
```

```python
options = Options()

options.add_argument("--lang=en")
options.add_argument('profile-directory=Default')

prefs = {
     "translate_whitelists": {"vi": "en"},
     "translate":{"enabled": "true"}
}
options.add_experimental_option("prefs", prefs)

cService = webdriver.ChromeService(executable_path="..\chromedriver.exe")
```

```python
df = pd.read_csv(r"itviec_url.csv")
```

```python
def crawler(url):
    try:
        driver = webdriver.Chrome(service = cService, options = options)
        driver.get(url)
        time.sleep(5)

        job_title = get_job_title(driver)
        company_name = get_company_name(driver)
        location = get_location(driver)
        working_time = get_time()
        requirement = get_requirement(driver)
        salary = get_salary(driver)
        technical_skill = get_technical_skill(driver)
        image_link = get_image_link(driver)
        job_url = url 

        driver.quit()
        
        return job_title, company_name, location, working_time, requirement, salary, technical_skill, image_link, job_url
    except: 
        return None*9
        
```

```python
def append_to_existing_csv(data, file_name):
    with open(file_name, mode='a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(data)
```

```python
existing_file = r"ITViec_data.csv"
```

```python
for row in df['link']:
    values = crawler(row)
    append_to_existing_csv(values, existing_file)
```
