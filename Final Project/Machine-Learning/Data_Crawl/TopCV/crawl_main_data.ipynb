```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time 
import csv
```

```python
def get_job_title(driver) -> str:
    try:
        title = driver.find_element(By.CLASS_NAME, "job-detail__info--title ")
        job_title = title.text

        return job_title
    except: 
        return None
```

```python
def get_company_name(driver) -> str:
    try:
        name = driver.find_element(By.CLASS_NAME, "company-name-label")
        Name = name.text    
        
        return Name
    except: 
        return None 
```

```python
def get_requirement(driver) -> list:
    try:
        requirements = []
        list_elements = driver.find_elements(By.XPATH, "/html/body/div[9]/div[3]/div[3]/div/div[1]/div[2]/div[1]/div[1]/div/div[2]/div")
        
        for element in list_elements:
            requirements.append(element.text)
        
        return requirements
    except: 
        return None 
    

# /html/body/div[9]/div[3]/div[3]/div/div[1]/div[2]/div[1]/div[1]/div/div[2]/div
# /html/body/div[9]/div[3]/div[3]/div/div[1]/div[2]/div[1]/div[1]/div/div[2]/div
```

```python
def get_salary(driver) -> str: 
    try:
        salary = driver.find_element(By.CLASS_NAME, "job-detail__info--section-content-value")
        
        if salary.text == 'The deal':
            sal =  'Negotiable'
        else:
            sal = salary.text

        return sal 
    except: 
        return None 
        
```

```python
def get_image_link(driver) -> str:
    try:
        image = driver.find_element(By.XPATH, "//div[@class='job-detail__company--information-item company-name']//img")
        link = image.get_attribute("src")
        
        return link
    except:
        return None
```

```python
def get_time(driver) -> str: 
    try: 
        wtime = driver.find_element(By.XPATH, "/html/body/div[9]/div[3]/div[3]/div/div[1]/div[2]/div[1]/div[1]/div/div[5]/div/div")
        wotime = wtime.text
        return wotime 
    except:
        return None 
```

```python
# Open Chrome

options = Options()

options.add_argument("--lang=en")
prefs = {
     "translate_whitelists": {"vi": "en"},
     "translate":{"enabled": "true"}
}

options.add_experimental_option("prefs", prefs)
cService = webdriver.ChromeService(executable_path="..\chromedriver.exe")
```

```python
def crawler(url):
    try:
        driver = webdriver.Chrome(service = cService, options = options)
        driver.get(url)
        time.sleep(0.5)
        job_title = get_job_title(driver)
        company_name = get_company_name(driver)
        requirement = get_requirement(driver)
        salary = get_salary(driver)
        # technical_skill = get_technical_skill(driver)
        # location = get_location(driver)
        # working_time = get_time(driver)
        image_link = get_image_link(driver)
        job_url = url 
        driver.quit()
        
        return [job_title, company_name, requirement, salary, image_link, job_url]
    except: 
        return [None]*6
        
```

```python
def append_to_existing_csv(data, file_name):
    with open(file_name, mode='a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(data)
```

```python
existing_file = r"ex_demo.csv"
```

```python
for row in df['link']:
    values = crawler(row)
    append_to_existing_csv(values, existing_file)
```
