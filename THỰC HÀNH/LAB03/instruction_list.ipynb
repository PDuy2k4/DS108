## 1) Tải dữ liệu về và đọc dữ liệu. 

### Tải và giải nén Dữ Liệu

```python
import requests
import zipfile
```

```python
zip_url = "https://d396qusza40orc.cloudfront.net/getdata%2Fprojectfiles%2FUCI HAR Dataset.zip"
zip_file_name  ="./human_activity.zip"
```

```python
response = requests.get(zip_url)
```

```python
if response.status_code == 200:
    with open(zip_file_name, "wb") as f:
        f.write(response.content)
```

```python
extract_dir = "./human_activity"
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)
```

### Đọc dữ liệu từ file zip

```python
import numpy as np
import pandas as pd
```

#### Training data

```python
X_train=np.loadtxt(extract_dir + "/UCI HAR Dataset/train/X_train.txt")
y_train=np.loadtxt(extract_dir + "/UCI HAR Dataset/train/y_train.txt",dtype=int)
subject_train=np.loadtxt(extract_dir + "/UCI HAR Dataset/train/subject_train.txt",dtype=int)
```

#### Testing data

```python
X_test=np.loadtxt(extract_dir + "/UCI HAR Dataset/test/X_test.txt")
y_test=np.loadtxt(extract_dir + "/UCI HAR Dataset/test/y_test.txt",dtype=int)
subject_test=np.loadtxt(extract_dir + "/UCI HAR Dataset/test/subject_test.txt",dtype=int)
```

### Gán tên cho các cột dữ liệu cho tập X

```python
features=np.loadtxt(extract_dir + "/UCI HAR Dataset/features.txt", dtype=str)
features=features[:,1] # get the feature names
```

```python
X_train=pd.DataFrame(X_train, columns=features)
X_test=pd.DataFrame(X_test, columns=features)
```

**Hiện thị thông tin của X khi đã gán tên**

```python
X_train.head()
```

```python
X_test.head()
```

## 3) Merge bộ dữ liệu train và test lại với nhau.

### Đặt tên cho tập y

```python
y_train=pd.DataFrame(y_train, columns=["Activity"])
y_test=pd.DataFrame(y_test, columns=["Activity"])
```

### Đặt tên cho tập subject

```python
subject_test=pd.DataFrame(subject_test, columns=["Subject"])
subject_train=pd.DataFrame(subject_train, columns=["Subject"])
```

### Merge dữ liệu test

```python
X_test_data=pd.concat([X_test,y_test,subject_test], axis=1)
```

### Merge dữ liệu train

```python
X_train_data=pd.concat([ X_train,y_train,subject_train], axis=1)
```

### Merge dữ liệu giữa tập train và test lại với nhau.

```python
X_data=pd.concat([X_train_data, X_test_data], axis=0).reset_index(drop=True)
X_data
```

## 4) Với các thuộc tính dữ liệu, chỉ lấy ra các thuộc tính giá trị trung bình (mean) và giá trị độ lệch chuẩn (standard deviation).

### Viết hàm lọc ra các cột có tên chứa từ "mean" hoặc "std"

```python
import re
```

```python
def filter(feature):
    return bool(re.match(r".*(mean|std).*", feature))
```

### Những feature khi đã lọc ra

```python
new_features=features[list(map(filter, features.copy()))]
new_features[:10] # lấy 10 phần tử đầu tiên
```

### Thêm cột Activity và Subject vào dữ liệu đã lọc

```python
new_features=np.append(new_features,["Activity","Subject"])
```

### X_data_new

```python
X_data_new=X_data[new_features]
X_data_new.head()
```

## 5) Đặt tên header cho bộ dữ liệu sao cho tên có ý nghĩa.

**Pattern**

```python
patterns = {
    r'^t': 'Time domain:',
    r'^f': 'Frequency domain:',
    r'-': ',',
    r'mean.*(.*)': 'mean value',
    r'std.*(.*)': 'standard deviation value',
    r'X': 'on X axis',
    r'Y': 'on Y axis',
    r'Z': 'on Z axis',
    r'AccJerk': 'acceleration jerk',
    r'Acc': 'acceleration',
    r'GyroJerk': 'angular velocity jerk',
    r'Gyro': 'angular velocity',
    r'Mag': 'magnitude'
}
```

**Viết hàm ánh xạ tên sau đó duyệt qua các tên cột của dữ liệu và ánh xạ**

```python
def map_column_name(column_name, patterns):
    for pattern, description in patterns.items():
        column_name = re.sub(pattern, description, column_name)
    return column_name
```

**Ánh xạ dữ liệu**

```python
X_data_new=X_data_new.rename(columns=lambda x: map_column_name(x, patterns))
X_data_new.head()
```

## 6) Tính giá trị trung bình của từng giá trị các thuộc tính trong bộ dữ liệu theo 2 thuộc tính là Activity và Subject

```python
ID_Activity=X_data_new.drop(columns=["Subject"]).groupby(["Activity"],as_index=True).mean() # Do hàm drop thực hiện ko có inplace nên không ảnh hưởng đến dữ liệu
ID_Activity.head()
```

```python
ID_Subject=X_data_new.drop(columns=["Activity"]).groupby(["Subject"],as_index=True).mean()
ID_Subject.head()
```

## 7) Đổi tên các biến Activity.

### Chuyển thành dạng dictionary để ánh xạ

```python
pattern_activity = dict(np.loadtxt(extract_dir + "/UCI HAR Dataset/activity_labels.txt",dtype=str))
pattern_activity
```

```python
X_data_new["Activity"]=X_data_new["Activity"].apply(lambda x: pattern_activity[str(x)])
X_data_new
```
