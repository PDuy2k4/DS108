## d. In ra số dữ liệu về ca lây nhiễm nhiều nhất trong khoảng tháng 01 và 02 tại Việt Nam (Lấy năm 2021).

```python
def findMaxConfirmed(x:pd.DataFrame())->pd.DataFrame():
    startTime=pd.Timestamp(2021, 1, 1)
    endTime=pd.Timestamp(2021, 2, 28)
    return x[x.Confirmed==x[(x.ObservationDate >=startTime) & (x.ObservationDate<=endTime)]["Confirmed"].max()]
```

```python
findMaxConfirmed(coronaVietnam)
```

```python
coronaPhilipine=coronaData[coronaData["Country/Region"]=="Philippines"]
findMaxConfirmed(coronaPhilipine)
```

## i. *Có nhận xét gì về số ca nhiễm mới tại Việt Nam giữa tháng 05/2020 và tháng 05/2021. Vẽ biểu đồ đường thể hiện số ca nhiễm mới trong 2 tháng trên.

```python
import matplotlib.pyplot as plt
```

### Biểu đồ số ca nhiễm tại Việt Nam vào tháng 5/2020

```python
y=coronaVietnam[(coronaVietnam.ObservationDate.dt.month==5)&(coronaVietnam.ObservationDate.dt.year==2020)]["Confirmed"]
x=coronaVietnam[(coronaVietnam.ObservationDate.dt.month==5)&(coronaVietnam.ObservationDate.dt.year==2020)]["ObservationDate"]


plt.figure(figsize=(12,6))
plt.plot(x,y)
plt.title("Số ca nhiễm tại Việt Nam vào tháng 5/2020")
plt.xlabel("Date")
plt.ylabel("Confirmed")
plt.show()
```

### Biểu đồ số ca nhiễm tại Việt Nam vào tháng 5/2020

```python
y=coronaVietnam[(coronaVietnam.ObservationDate.dt.month==5)&(coronaVietnam.ObservationDate.dt.year==2021)]["Confirmed"]
x=coronaVietnam[(coronaVietnam.ObservationDate.dt.month==5)&(coronaVietnam.ObservationDate.dt.year==2021)]["ObservationDate"]


plt.figure(figsize=(12,6))
plt.plot(x,y)
plt.title("Số ca nhiễm tại Việt Nam vào tháng 5/2021")
plt.xlabel("Date")
plt.ylabel("Confirmed")
plt.show()
```

## k.Vẽ biểu đồ về số ca lây nhiễm nhiều nhất của 3 quốc gia: Vietnam, Indonesia và Philippine trong 2 tháng gồm 01 và tháng 02 năm 2021.

### Lọc dữ liệu

```python
import seaborn as sns
filtered_data=coronaData[(coronaData.ObservationDate.dt.month.isin([1,2]))&(coronaData.ObservationDate.dt.year==2021)&(coronaData["Country/Region"].isin(["Vietnam","Indonesia","Philippines"]))].groupby(["Country/Region",pd.Grouper(key='ObservationDate', freq='M')],as_index=False).max("Confirmed")
filtered_data
```

### Vẽ đồ thị

```python
plt.figure(figsize=(12,6))
sns.barplot(data=filtered_data,x="Confirmed",y="ObservationDate",hue="Country/Region")
plt.title("Số ca nhiễm tối đa của 3 quốc gia vào tháng 1 và 2 năm 2021")
plt.yticks(ticks=[0, 1], labels=["tháng 1", "tháng 2"])
plt.xlabel("Số ca nhiễm tối đa")
plt.ylabel("Tháng")
plt.xscale('log')
plt.show()
```
